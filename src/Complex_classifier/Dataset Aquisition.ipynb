## Taking Photos with a Webcam

```python
def take_photos(folder_name, delay_s=3, log_out = True):
    folder_path = f"../../data/complex/{folder_name}"
    #os.makedirs(folder_path, exist_ok=True)

    while delay_s > 0:
        #print(f"{delay_s}...")
        time.sleep(1)
        delay_s -= 1

    # Capture RAW image in PPM format (lossless)
    raw_image_path = f"{folder_path}.jpg"
    command = f"fswebcam --no-banner {raw_image_path}"
    subprocess.run(command, shell=True)
```

```python
i = 0

for i in range(0, 150):
    take_photos(f"me/me_{i+1}", 2)
```
